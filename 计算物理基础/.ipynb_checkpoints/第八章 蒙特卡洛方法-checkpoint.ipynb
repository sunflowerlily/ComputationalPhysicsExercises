8.1 使用舍选抽样法产生正态分布的随机数

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_pdf(x):
    """
    标准正态分布的概率密度函数
    """
    return np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)

def rejection_sampling(n):
    """
    使用舍选抽样法生成正态分布随机数
    
    参数:
    n -- 生成随机数的数量
    
    返回:
    正态分布随机数数组
    """
    samples = []
    while len(samples) < n:
        x = np.random.uniform(-3, 3)
        y = np.random.uniform(0, 1)
        if y <= normal_pdf(x):
            samples.append(x)
    return np.array(samples)

# 生成正态分布随机数
n = 10000
samples = rejection_sampling(n)

# 绘制直方图
plt.figure(figsize=(8, 4))
plt.hist(samples, bins=50, density=True, alpha=0.7, label='Samples')
x = np.linspace(-3, 3, 100)
plt.plot(x, normal_pdf(x), 'r-', label='PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution (Rejection Sampling)')
plt.legend()
plt.show()
```

8.2 用MC方法计算单位直径的小球体积，取10次计算的结果平均计算均方差

```python
import numpy as np

def estimate_sphere_volume(num_points):
    """
    使用蒙特卡洛方法估计单位直径小球的体积
    
    参数:
    num_points -- 随机点的数量
    
    返回:
    小球体积的估计值
    """
    inside_points = 0
    for _ in range(num_points):
        x, y, z = np.random.uniform(-1, 1, size=3)
        if x**2 + y**2 + z**2 <= 1:
            inside_points += 1
    return 8 * (inside_points / num_points)

# 设置参数
num_experiments = 10
num_points = 100000

# 进行10次独立的实验
results = []
for _ in range(num_experiments):
    volume = estimate_sphere_volume(num_points)
    results.append(volume)

# 计算均值和均方差
mean_volume = np.mean(results)
std_dev = np.std(results, ddof=1)
print(f"Mean volume: {mean_volume:.4f}")
print(f"Standard deviation: {std_dev:.4f}")
```

8.11.3 模仿教材中的做法验证中心极限定理。要求取10个随机数相加，并用正态分布函数来拟合。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 生成10组,每组包含10个随机数
n_groups = 10000
group_size = 10
groups = np.random.rand(n_groups, group_size)

# 计算每组随机数的和
group_sums = np.sum(groups, axis=1)

# 计算理论均值和标准差
theoretical_mean = group_size / 2
theoretical_std = np.sqrt(group_size / 12)

# 绘制直方图和理论正态分布曲线
plt.figure(figsize=(8, 4))
plt.hist(group_sums, bins=50, density=True, alpha=0.7, label='Group Sums')
x = np.linspace(group_sums.min(), group_sums.max(), 100)
pdf = norm.pdf(x, loc=theoretical_mean, scale=theoretical_std)
plt.plot(x, pdf, 'r-', label='Theoretical Normal Distribution')
plt.xlabel('Sum of 10 Random Numbers')
plt.ylabel('Density')
plt.title('Central Limit Theorem Verification')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit

# 生成10组,每组包含10个随机数
n_groups = 10000
group_size = 10
groups = np.random.rand(n_groups, group_size)

# 计算每组随机数的和
group_sums = np.sum(groups, axis=1)

# 定义正态分布函数
def normal_dist(x, mean, std):
    return norm.pdf(x, loc=mean, scale=std)

# 拟合正态分布曲线
hist, bin_edges = np.histogram(group_sums, bins=50, density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
popt, _ = curve_fit(normal_dist, bin_centers, hist)
fitted_mean, fitted_std = popt

# 绘制直方图和拟合的正态分布曲线
plt.figure(figsize=(8, 4))
plt.hist(group_sums, bins=50, density=True, alpha=0.7, label='Group Sums')
x = np.linspace(group_sums.min(), group_sums.max(), 100)
pdf_fitted = normal_dist(x, *popt)
plt.plot(x, pdf_fitted, 'r-', label='Fitted Normal Distribution')
plt.xlabel('Sum of 10 Random Numbers')
plt.ylabel('Density')
plt.title('Central Limit Theorem Verification')
plt.legend()
plt.show()

print(f"Fitted mean: {fitted_mean:.4f}")
print(f"Fitted standard deviation: {fitted_std:.4f}")
```

### 8.11-4 
用平均值法和辅助函数法计算积分
$$I=\int_0^1 e^x da$$
分别取5次计算结果计算均方差，比较二者的精度。

#### 1. 平均值法

```python
import numpy as np

def average_method(n_samples):
    x = np.random.uniform(0, 1, n_samples)
    return np.mean(np.exp(x))

n_samples = 1000000
results_average = [average_method(n_samples) for _ in range(5)]
mean_average = np.mean(results_average)
std_average = np.std(results_average)

print("平均值法:")
print(f"积分估计值: {mean_average:.6f}")
print(f"均方差: {std_average:.6f}")
```

#### 2. 辅助函数法

```python
import numpy as np

def auxiliary_method(n_samples):
    x = np.random.uniform(0, 1, n_samples)
    return (np.exp(1) - 1) * np.mean(x <= np.exp(x) / np.exp(1))

n_samples = 1000000
results_auxiliary = [auxiliary_method(n_samples) for _ in range(5)]
mean_auxiliary = np.mean(results_auxiliary)
std_auxiliary = np.std(results_auxiliary)

print("辅助函数法:")
print(f"积分估计值: {mean_auxiliary:.6f}")
print(f"均方差: {std_auxiliary:.6f}")
```

#### 5
用蒙特卡洛面积法计算
$$I=\int_0^2 (4-x^2)dx$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 4 - x**2

def monte_carlo_integration(n_samples):
    x_min, x_max = 0, 2
    y_min, y_max = 0, 4
    
    x = np.random.uniform(x_min, x_max, n_samples)
    y = np.random.uniform(y_min, y_max, n_samples)
    
    below_curve = (y <= f(x)).sum()
    total_points = n_samples
    
    area = (x_max - x_min) * (y_max - y_min) * below_curve / total_points
    return area

n_samples = 100000
integral_estimate = monte_carlo_integration(n_samples)

print(f"积分的估计值: {integral_estimate:.4f}")

x = np.linspace(0, 2, 100)
plt.figure(figsize=(8, 4))
plt.plot(x, f(x), 'b-', label='f(x) = 4 - x^2')
plt.fill_between(x, f(x), color='lightskyblue', alpha=0.5)
plt.xlim(0, 2)
plt.ylim(0, 4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Monte Carlo Area Method')
plt.legend()
plt.show()
```

#### 8.11-6
用MC方法模拟伽尔顿板实验。

思路是：假设有一个钉子组成的等间距的垂直阵列，小珠从上部滚落下来，碰到一个钉子后将等概率向左或向右运动。为了简单，还要假定，向左运动的小珠将碰到下一排左边一格的钉子，向右运动的钉子将碰到下一排右边一格的钉子。小珠一直这样落下去，直到底部。用模拟证明，即使所有小珠都从一个位置下落，在底部也会形成正态分布。

```python
import numpy as np
import matplotlib.pyplot as plt

def galton_board(n_rows, n_balls):
    final_positions = np.zeros(n_rows + 1, dtype=int)
    
    for _ in range(n_balls):
        position = 0
        for _ in range(n_rows):
            if np.random.rand() < 0.5:
                position -= 1
            else:
                position += 1
        final_positions[position] += 1
    
    return final_positions

n_rows = 10
n_balls = 10000

positions = galton_board(n_rows, n_balls)

x = np.arange(-n_rows, n_rows + 1, 2)
plt.figure(figsize=(8, 4))
plt.bar(x, positions, width=0.8, align='center', alpha=0.7)
plt.xlim(-n_rows - 1, n_rows + 1)
plt.xlabel('Final Position')
plt.ylabel('Number of Balls')
plt.title('Galton Board Simulation')
plt.show()
```

#### 8.11-9.
画出如图8.21所示的IFS分形树冠，它的迭代法则如下，作图的出发点
可取在（0，0）处。后面的百分数表示概率。
$$
\begin{pmatrix}
x_{n+1}\\
y_{n+1}
\end{pmatrix}=\begin{cases}
\begin{pmatrix}
0.01x_{n}\\
0.45y_{n}
\end{pmatrix}&5\%\\
\begin{pmatrix}
-0.01x_{n}\\
-0.45y_{n}+0.2
\end{pmatrix}&15\%\\
\begin{pmatrix}
0.42x_{n}-0.42y_n\\
0.42x_n+0.42y_{n}+0.2
\end{pmatrix}&40\%\\
\begin{pmatrix}
0.42x_{n}+0.42\\
-0.42x_n+0.42y_{n}+0.2
\end{pmatrix}&40\%\\
\end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50000
x = np.zeros(N)
y = np.zeros(N)
x[0], y[0] = 0, 0

for i in range(1, N):
    r = np.random.rand()
    if r < 0.05:
        x[i] = 0.01 * x[i-1]
        y[i] = 0.45 * y[i-1]
    elif r < 0.2:
        x[i] = -0.01 * x[i-1]
        y[i] = -0.45 * y[i-1] + 0.2
    elif r < 0.6:
        x[i] = 0.42 * x[i-1] - 0.42 * y[i-1]
        y[i] = 0.42 * x[i-1] + 0.42 * y[i-1] + 0.2
    else:
        x[i] = 0.42 * x[i-1] + 0.42 * y[i-1]
        y[i] = -0.42 * x[i-1] + 0.42 * y[i-1] + 0.2

plt.figure(figsize=(6, 8))
plt.plot(x, y, ',', color='green', markersize=1)
plt.axis('equal')
plt.axis('off')
plt.tight_layout()
plt.show()
```

#### 8-11-10
利用迭代可以画出山水图、烟雾、火焰等美妙的图形，也可以作出分形的动画，尝试一下从网上找到类似的程序或者自已编写这类程序，体会一下数学的神奇。

```python
import numpy as np
import matplotlib.pyplot as plt

def midpoint_displacement(start, end, roughness, vertical_displacement=None, num_iterations=16):
    if vertical_displacement is None:
        vertical_displacement = (end - start) / 2

    points = [start, end]
    iteration = 1

    while iteration <= num_iterations:
        points_tup = tuple(points)
        for i in range(len(points_tup) - 1):
            midpoint = (points_tup[i] + points_tup[i + 1]) / 2
            midpoint += (np.random.random() - 0.5) * vertical_displacement
            points.append(midpoint)
        points.sort()
        vertical_displacement *= 2 ** (-roughness)
        iteration += 1

    return points

def draw_landscape(width, height, roughness):
    x = midpoint_displacement(0, width, roughness)
    y = midpoint_displacement(0, height, roughness)

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(x, y, 'k')
    ax.fill_between(x, y, 0, color='g')
    ax.set_xlim(0, width)
    ax.set_ylim(0, height)
    ax.axis('off')
    plt.tight_layout()
    plt.show()

# 设置图形的宽度、高度和粗糙度
width = 1000
height = 500
roughness = 0.5

# 绘制分形山水图
draw_landscape(width, height, roughness)
```

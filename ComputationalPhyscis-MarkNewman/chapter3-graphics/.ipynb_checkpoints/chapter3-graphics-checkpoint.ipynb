```python
from visual import sphere, color, rate

# Define the radii of the planets and the Sun (in millions of km)
radii = {
    'Mercury': 2440,
    'Venus': 6052,
    'Earth': 6371,
    'Mars': 3386,
    'Jupiter': 69173,
    'Saturn': 57316,
    'Sun': 695500
}

# Define the radii of the orbits (in millions of km) and the periods of orbit (in days)
orbits = {
    'Mercury': 57.9,
    'Venus': 108.2,
    'Earth': 149.6,
    'Mars': 227.9,
    'Jupiter': 778.5,
    'Saturn': 1433.4
}

# Create spheres for the Sun and planets
spheres = {}
for planet, radius in radii.items():
    if planet == 'Sun':
        spheres[planet] = sphere(pos=(0, 0, 0), radius=radius*0.1, color=color.yellow)
    else:
        spheres[planet] = sphere(pos=(orbits[planet], 0, 0), radius=radius*0.0005, color=color.blue)

# Animate the motion of the planets
t = 0
dt = 1  # Time step
while True:
    rate(30)  # Adjust the speed of the animation
    for planet, radius in radii.items():
        if planet != 'Sun':
            spheres[planet].pos.x = orbits[planet] * cos(2 * pi * t / 365)
    t += dt
```

**Deterministic chaos and the Feigenbaum plot**

最著名的混沌现象的例子是*Logistic*模型，是一个描述自然界生物种群中昆虫数目变化的数学模型，由下面的方程确定
$$
x' = rx(1-x).
$$
对于给定常数$r$，你可以选一个$x$的值，比如取$x=\dfrac{1}{2}$, 之后代入方程的右手边，计算出一个新的$x$，再将新的$x$代入，不断迭代.

    1. 当$r=2$时，$x$趋于0.5. 结论:没有分岔.

    2. 当$r=3.2$时, $x$趋于0.51304，0.79946. 结论: 周期2分岔.

    3. 当$r=3.45$时, 周期4分岔.

#### 练习 3.7：曼德博集
曼德博集合以其发现者、法国数学家本诺伊特·曼德博命名，是一种分形结构，是一个可以无限细分的数学对象，结构内包含结构，可以一直不断细分下午。曼德博集合的定义涉及复数，如下所述。考虑方程 
$$z' = z^2 + c,$$
其中$ z $是复数，$c $是复常数。对于任何给定的$ c $值，这个方程将输入数字$ z $转换为输出数字$ z'$。曼德博集合的定义涉及这个方程的重复迭代：我们取一个初始起始值$ z$，将其输入方程得到新的值$ z'$。然后我们取那个值再次输入得到另一个值，如此反复。曼德博集合是满足以下定义的复平面上的点集：

*对于给定的复数$ c $值，从$ z = 0 $开始重复迭代。如果得到的值的幅度$ |z| $大于$ 2$，则复平面上位置$ c $的点不包含在曼德博集合中，否则就包含在集合中。*

原则上，为了使用这个定义，人们将不得不无限次迭代以证明一个点在曼德博集合中，因为只有当迭代从未通过$ |z| = 2$ 时，点才在集合中。然而，在实践中，人们通常只执行一定数量的迭代，比如说 100 次，如果到那时$ |z| $还没有超过 $2$，那么我们认为这足够好了。

编写一个程序，通过在$ N \times N $网格上对所有$ c = x + iy $的值进行迭代，制作曼德博集合的图像，覆盖区域 $-2\le x\le 2$ 和 $-2\le y\le 2$。制作一个密度图，其中曼德博集合内的网格点被涂成黑色，外面的被涂成白色。曼德博集合有一个非常独特的形状，看起来像一个长鼻子的甲虫——你看到它时就会知道。

提示：你可能会发现从相当粗糙的网格开始很有用，即小的$ N$值——比如$ N = 100$——这样你的程序在测试时运行得很快。一旦你确信它正确无误，就增加$ N $的值以产生集合形状的最终高质量图像。

如果你感到有意思，这里有一个相同练习的另一个变体，可以产生惊人的图片。不是简单地将点涂成黑色或白色，而是根据方程迭代次数（在$ |z| $大于 2 之前，或者如果$ |z| $从未大于 2，则为最大迭代次数）来着色。如果你使用 Python 提供的更丰富多彩的密度图颜色方案，比如“热”或“喷气”方案，你可以通过这种方式制作一些壮观的图像。另一个有趣的变体是根据迭代次数的对数来着色，这有助于揭示集合外的一些更精细的结构。

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
N = 1000  # Grid size
max_iterations = 100  # Maximum number of iterations
xmin, xmax, ymin, ymax = -2, 2, -2, 2  # Region of the complex plane

# Create an array to store the Mandelbrot set
mandelbrot_set = np.zeros((N, N))

# Create a grid of complex numbers
x = np.linspace(xmin, xmax, N)
y = np.linspace(ymin, ymax, N)
X, Y = np.meshgrid(x, y)
c = X + 1j * Y

# Iterate the Mandelbrot equation
z = c
for i in range(max_iterations):
    z = z**2 + c
    mask = np.abs(z) < 2
    mandelbrot_set += mask

# Create the density plot
plt.imshow(mandelbrot_set, extent=(xmin, xmax, ymin, ymax), cmap='binary', origin='lower')
plt.colorbar()
plt.show()
```

### Exercise 3.7
**The Mandelbrot set**

The Mandelbrot set, named after its discoverer, the French
mathematician Beno\^{\i}t Mandelbrot, is a \defn{fractal}, an infinitely
ramified mathematical object that contains structure within structure
within structure, as deep as we care to look.  The definition of the
Mandelbrot set is in terms of complex numbers as follows.

Consider the equation
$$
z' = z^2 + c,
$$
where $z$ is a complex number and $c$ is a complex constant.  For any given
value of $c$ this equation turns an input number $z$ into an output
number $z'$.  The definition of the Mandelbrot set involves the repeated
iteration of this equation: we take an initial starting value of $z$ and
feed it into the equation to get a new value $z'$.  Then we take that value
and feed it in again to get another value, and so forth.  The Mandelbrot
set is the set of points in the complex plane that satisfies the following
definition:

 *For a given complex value of $c$, start with $z=0$ and iterate
repeatedly.  If the magnitude $|z|$ of the resulting value is ever greater
than 2, then the point in the complex plane at position $c$ is \emph{not}
in the Mandelbrot set, otherwise it is in the set.*

In order to use this definition one would, in principle, have to iterate

### 最小二乘拟合和光电效应

物理实验中产生的数据常常展现出如下所示的接近线性关系

<img src="leastsq.png" alt="leastsq" title="leastsq" width=200>

这里的实线代表潜在的线性关系，点代表测量的数据。


直线可以用$y=mx+c$描述，可以用最小二乘法来确定合适的斜率$m$和截距$c$。

假设我们猜测一条直线的$m$和$c$的值.然后计算数据点到这条直线的竖直距离，如图中的竖直线段所示，之后我们计算这些距离的平方和, 用$\chi^2$表示，如果共有$N$个数据点,坐标分别为$(x_i,y_i)$, 则有

$$
\chi^2 = \sum_{i=1}^N (mx_i+c-y_i)^2.
$$

对直线的最小二乘拟合就是找到令$\chi^2$取得最小值的直线。可以通过对$m$和$c$分别求导数并让导数为零来确定最小值的点：

\begin{align*}
m \sum_{i=1}^N x_i^2 + c \sum_{i=1}^N x_i - \sum_{i=1}^N x_iy_i &= 0, \\
m \sum_{i=1}^N x_i + cN - \sum_{i=1}^N y_i &= 0.
\end{align*}

为方便起见，定义下列变量:

$$
E_x = {1\over N} \sum_{i=1}^N x_i,\qquad
E_y = {1\over N} \sum_{i=1}^N y_i,\qquad
E_{xx} = {1\over N} \sum_{i=1}^N x_i^2,\qquad
E_{xy} = {1\over N} \sum_{i=1}^N x_iy_i,
$$
所以方程可以简化为

\begin{align*}
mE_{xx} + cE_x &= E_{xy}\,, \\
mE_x + c &= E_y\,.
\end{align*}
解这个方程组可以得到$m$和$c$
$$
m = {E_{xy}-E_x E_y\over E_{xx} - E_x^2},\qquad
c = {E_{xx}E_y-E_x E_{xy}\over E_{xx} - E_x^2}.
$$


1. 本题提供了一个数据文件millikan.txt. 这个文件包含两列数值，分别代表数据点的$x$和$y$坐标. 写程序画数据点，用点或圆表示每个数据点.  

2. 在你的程序中添加代码，在画图之前先计算$E_x$, $E_y$, $E_{xx}$, 和 $E_{xy}$，之后计算并打印$m$和$c$的值。

3. 用这组$m$和$c$的数值，利用关系$ m x_i + c$生成新的一组$y$，对应直线上的数据点，并将这些数据点以直线的形式和原来的数据点画在一张图上. 你最终应该得到一个数据点加上穿过它们的直线的图。

4. 文件millikan.txt中的数据来源于Robert Millikan所做的著名的光电效应的试验. 当合适波长的光子打在金属表面时，金属表面的电子可能会因吸收能量而释放到自由空间中.被释放的电子能量等于打在上面的光子能量减去一个小量$\phi$, 代表电子从表面逃逸所需要的能量. 光子能量为$h\nu$，$h$代表普朗克常量,$\nu$代表光的频率.逃逸电子的能量可以通过使电子停止的电压$V$来测量，有如下关系:
$$
V = {h\over e}\nu - \phi,
$$
这里$e$是电子电荷. 这个方程是1905年由爱因斯坦首先给出的。

millikan.txt中的数据第一列代表$\nu$, 单位是赫兹,第二列代表电压$V$, 单位是V. 使用上面的方程和你的程序计算普朗克常量（电子电荷取$1.602\times10^{-19}\,$C）,并与实际值相比较.

这个计算基本上与Millikan自己做的测量普朗克常量的试验相同，尽管由于缺少计算机，他是用眼睛拟合的直线. 这个工作是Millikan获得1923年的诺贝尔物理奖工作的一部分.

```python
#1. 读入实验数据
data = np.loadtxt('millikan.txt')
x = data[:,0]
y = data[:,1]
plt.plot(x,y,'ko')
```

```python
#使用最小二乘法得到m和c的值
Ex=x.mean()
Ey=y.mean()
Exx=np.mean(x*x)
Exy = np.mean(x*y)
m = (Exy-Ex*Ey)/(Exx-Ex*Ex)
c = (Exx*Ey - Ex*Exy)/(Exx-Ex*Ex)
print('m = ', m,'c = ',c)
```

```python
# 3. 将最小二乘直线和数据点画在图上
xt = np.linspace(x.min(),x.max(),100)
yt = m*xt + c
plt.plot(x,y,'ko')
plt.plot(xt,yt)
```

```python
# 4. 计算普朗克常量，并与实际值比较
from scipy.constants import e,h
print(m*e,h)
```

#### 另外一种解法
此题也可以用线性代数中的法方程求解, 系数矩阵为
$$
A=\left[\begin{matrix}
1&x_1\\
1&x_2\\
\vdots&\vdots\\
1&x_n
\end{matrix}
\right]
$$
未知参数矩阵为
$$
\vec{x}=\left[\begin{matrix}
c\\
m
\end{matrix}
\right]
$$
常数矩阵为
$$
\vec{b}=\left[\begin{matrix}
y_1\\
y_2\\
\vdots\\
y_n
\end{matrix}
\right]
$$
矩阵的法方程为
$$A^TA\vec{x}=A^T\vec{b}$$
解上面的方程，得到最小二乘解

```python
X=np.hstack((np.ones(x.size).reshape(x.size,1),x.reshape(x.size,1)))
c,m = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,y))
print('c=',c,'m=',m)
```
